```python
# Write a lambda function that checks whether a number is even or odd.
no_check = lambda x : "even" if x % 2 == 0 else "odd"
print(no_check(5))
```

```python
# Write a lambda function that returns:

# "Adult" if age >= 18
# "Minor" otherwise
age_checker = lambda x : "Adult" if x >= 18 else "minor"
print(age_checker(10))
```

```python
# Convert a string into uppercase using lambda.
lower_uper = lambda x : x.upper()

print(lower_uper("gautam"))
```

```python
length = lambda x:len(x)
print(length("gautam"))

```
