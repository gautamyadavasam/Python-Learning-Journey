SHORTING OF LIST

```python
#Method 1 : Shorting from Asecding 
Numbers = [4,3,10,6,7,8,1]
Numbers.sort()
print(Numbers)
#It Short the List from asecding to desceding 
```

```python
#Method 2 :- shorting from  Descending to asceding 
Numbers = [4,3,10,6,7,8,1]
Numbers.sort(reverse=True)
print(Numbers)
#it Prints the numbers in a Descending Order 
```

COPYING OF A LIST

```python
#METHOD 1
# Slicing  Method
Numbers = [4,3,10,6,7,8,1]
New = Numbers[:]
print(New)
#Its copy a list and create a new list as simple as control c  print work as cntrl v
```

```python
#METHOD 2
#classic Copy Method
Numbers = [4,3,10,6,7,8,1]
New_list = Numbers.copy()
print(New_list)
```
