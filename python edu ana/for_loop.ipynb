Question:- Create a list called scores = [45, 80, 30, 95, 60]. Use a for loop to go through them, but only print the scores that are 70 or higher (passing scores)

```python
scores = [45, 80, 30, 95, 60]
count = 0 
for n in scores:
    if n >70:
        print(n)
        count= count+1
        print("total people who pased: ", count)
```

Skip any transaction that is negative (this is a mistake in the data).

Stop everything if you find a transaction over $500 (this is a "Fraud Alert").

Here is your starter code:

Python

processed_total = 0

for t in transactions:
    # 1. If 't' is less than 0, skip it using 'continue'
    
    # 2. If 't' is greater than 500, print "FRAUD!" and 'break'
    
    # 3. Otherwise, add 't' to 'processed_total' and print "Processed: " + t

```python
transactions = [10, 25, -5, 100, 600, 20, 15]
money_spent= []
for t in transactions:
    if t <0:
        print("This is a mistake in the data" , t )
        continue     

    if t >= 500:
        print("This is a fraud", t)
        break

    money_spent.append(t)

print("total Money Spent" , money_spent)
```
